```python
import matplotlib.pyplot as plt

# Data
labels = ['Fear of Job Loss', 'Fear of Misuse', 'Excited by Innovation', 'Trust in AI']
sizes = [40, 25, 20, 15]

# Plot
plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Public Perception of AI Adoption")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example Data
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
publications = [5000, 7000, 11000, 15000, 22000, 30000, 45000, 60000, 75000]

plt.figure(figsize=(10,6))
plt.plot(years, publications, marker='o')
plt.title("Growth of AI Research Publications")
plt.xlabel("Year")
plt.ylabel("Number of Publications")
plt.grid(True)
plt.show()
```
